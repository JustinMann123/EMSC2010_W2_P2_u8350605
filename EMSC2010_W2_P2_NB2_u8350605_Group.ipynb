# Fermi estimation problem: How many grains of sand in the top 1m of Bondi Beach on the average Low tide

### This code was written by Freddie, Clover, Justin, Ed

```python
Length = 1000 #Using m as our main unit, Freddie added this code
```

```python
Sand_Depth = 1 #We decided only to count the top 1 m
```

```python
Width = 70 #Width at low tide in m
```

```python
Sand_Diam = 0.0001 ## unit in m
```

```python
Sandperm = 1/Sand_Diam
Sandperm
```

```python
Sandperm2 = Sandperm**2
Sandperm2
```

```python
Sandperm3 = 0.95*(Sandperm2**2) #Nine Hundred and Fifty Trillion
Sandperm3
```

```python
Beach_Volume = Length*Width*Sand_Depth
Beach_Volume
```

## Final Sand Volume Without Flux

```python
Sand_Volume_Whole = Beach_Volume*Sandperm3 # Six sextillion, six hundred fifty quintillion.
Sand_Volume_Whole
```

#Flux Out Estimation

```python
Patrons_per_hour = 5000 #on an average day in summer
Patrons_per_day = Patrons_per_hour*12
Patrons_per_week = Patrons_per_day*7
Sand_Grains_Subtracted_per_patrons = 5000
Sand_taken_per_week = Sand_Grains_Subtracted_per_patrons*Patrons_per_week
Sand_taken_per_week
```

# Final Sand Volume

```python
Fin_Volume = Sand_Volume_Whole - Sand_taken_per_week
Fin_Volume
```

The final volume of sand is **Six hundred sixty-four quintillion, nine hundred ninety-nine quadrillion, nine hundred ninety-nine trillion, nine hundred ninety-seven billion, nine hundred million.**
